## Clean

#### Define

#### Code & Test

```python
df.columns = df.columns.str.lower()
df
```

- "screen_resolution", "access_type", "screen_type" columns should be derived from "screenresolution" column
                         
#### Define
- "screen_resolution"  ----- use RegeX to detect the pattern of resolution
-  "access_type"       ----- use find() to detect the word "Touchscreen" and if not found put non_touch
-  "screen_type"       ----- use find() to detect the word "IPS" and if not found put other
#### Code & Test

```python
df.insert(6,column ='screen_resolution', value ='')
df.insert(7,column ='access_type', value ='')
df.insert(8,column ='screen_type', value ='')
df
```

```python
df['screen_resolution'] = df['screenresolution'].str.findall("[0-9]+x[0-9]+").str[0]
```

```python
df['screen_resolution'].value_counts()
```

```python
touch = df['screenresolution'].str.findall("Touchscreen").str[0]
touch.value_counts()
df['access_type'] = touch
df['access_type'] = df['access_type'].fillna('Non-Touch')
df['access_type']
```

```python
ips = df['screenresolution'].str.findall("IPS").str[0]
ips
```

```python
df['screen_type'] = ips
df['screen_type'] = df['screen_type'].fillna('other')
df['screen_type']
```

```python
df
```

- Derive from Cpu column  "Brand", "core", "generation" columns

#### Define
- "Brand"  ---- use split to detect the first word 
- "core"   ---- use RegeX to detect the letter i and the number
- "generation" -- use RegeX to detect the number with the the codeing attached to it and then use slicing to get the 1st digit

#### Code & Test

```python
df.insert(10,column ='cpu_brand', value ='')
df.insert(11,column ='cpu_core', value ='')
df.insert(12,column ='cpu_generation', value ='')
df
```

```python
df['cpu_brand'] = df['cpu'].str.split(' ',-1).str[0]
df['cpu_brand'].value_counts()
```

```python
df['cpu_core'] = df['cpu'].str.findall("i[3-7]{1}").str[0]
df['cpu_core'] = df['cpu_core'].fillna('Other')
df['cpu_core'].value_counts()
```

```python
df.info()
```

```python
df['cpu_generation'] = df['cpu'].str.findall("[0-9]+[A-Z]").str[0]
df['cpu_generation'] = df['cpu_generation'].str.replace("[0-9]G",'Unknown')
df['cpu_generation'] = df['cpu_generation'].str.replace("[0-9]Unknown",'Unknown')
df['cpu_generation'] = df['cpu_generation'].str[0]
df['cpu_generation'] = df['cpu_generation'].str.replace("U",'Unknown')
df['cpu_generation'].value_counts()
```

```python
df
```

- memory_type (SSD, HDD, Flash ...etc), memory_capacity
#### Define
- memory_type ----- use RegeX to split on "xxxGB" 
#### Code & Test

```python
df['memory'].value_counts().sum()
```

```python
df.insert(15,column ='memory_type', value ='')
df.insert(16,column ='memory_capacity', value ='')
```

```python
df['memory_type1'] = df['memory'].str.split(" ",-1).str[1]
df['memory_type1'].value_counts()
```

```python
df['memory_type1'].isnull().sum()
```

```python
df['memory_type2'] = df['memory'].str.replace(" ","")
df['memory_type2'] = df['memory'].str.split("+",-1).str[1]
df['memory_type2'] = '/'+ ' '+ df['memory_type2'].str.split(" ",-1).str[-1]
df['memory_type2'].value_counts()
```

```python
df['memory_type2'].isnull().sum()
```

```python
df['memory_type2'] = df['memory_type2'].fillna('')
df['memory_type2'].isnull().sum()
```

```python
df['memory_type'] = df['memory_type1'] + ' ' + df['memory_type2'] 
```

```python
df['memory_type'].value_counts()
```

```python
df['memory_type'].value_counts().sum()
```

```python
df.drop(['memory_type1', 'memory_type2'], axis = 1, inplace = True)
```

```python
df
```

```python
df['memory_capacity1'] = df['memory'].str.split(" ",-1).str[0] + ' '
df['memory_capacity1'].value_counts()
```

```python
df['memory_capacity1'].value_counts().sum()
```

```python
df['memory_capacity2'] = df['memory'].str.split("+",-1).str[1].str[:7]
df['memory_capacity2'] = '/' +' '+ df['memory_capacity2'].str.replace('H','').str.lstrip()
df['memory_capacity2'].value_counts()
```

```python
df['memory_capacity2'].isnull().sum()
```

```python
df['memory_capacity2'] = df['memory_capacity2'].fillna('')
```

```python
df['memory_capacity'] = df['memory_capacity1'] + df['memory_capacity2']
df['memory_capacity'].value_counts()
```

```python
df.drop(['memory_capacity1', 'memory_capacity2'], axis = 1, inplace = True)
df
```
